**Q1.** This question provides some practice doing exploratory data analysis and visualization.

The "relevant" variables for this question are:
  - `level` - Level of institution (4-year, 2-year)
  - `aid_value` - The average amount of student aid going to undergraduate recipients
  - `control` - Public, Private not-for-profit, Private for-profit
  - `grad_100_value` - percentage of first-time, full-time, degree-seeking undergraduates who complete a degree or certificate program within 100 percent of expected time (bachelor's-seeking group at 4-year institutions)

1. Load the `./data/college_completion.csv` data with Pandas.

```python
# RESOURCES
# https://www.javatpoint.com/how-to-get-the-number-of-rows-and-columns-in-dataframe-python#:~:text=len()&text=The%20len()%20function%20can,columns%20are%20represented%20by%20dataframe.

import pandas as pd
college_completion = pd.read_csv ('college_completion.csv', low_memory=False)
```

2. What are are the dimensions of the data? How many observations are there? What are the variables included? Use `.head()` to examine the first few rows of data.

```python
rows = len(college_completion.axes[0])  
cols = len(college_completion.axes[1])  
print(f"Rows 'num of observations': {rows}")
print(f"Columns 'num of variables': {cols}\n")
print(college_completion.head())
```

3. Cross tabulate `control` and `level`. Describe the patterns you see.

```python
pd.crosstab(college_completion['control'], college_completion['level'])
```

4. For `grad_100_value`, create a histogram, kernel density plot, boxplot, and statistical description.

```python
# RESOURCES
# https://stackoverflow.com/questions/66922511/replace-all-inf-inf-values-with-nan-in-a-pandas-dataframe

import seaborn as sns
import numpy as np
college_completion = college_completion.replace([np.inf, -np.inf], np.nan)
sns.histplot(college_completion['grad_100_value'])
```

```python
sns.kdeplot(college_completion['grad_100_value'])
```

```python
sns.boxplot(college_completion['grad_100_value'])
```

```python
college_completion['grad_100_value'].describe()
```

5. For `grad_100_value`, create a grouped kernel density plot by `control` and by `level`. Describe what you see. Use `groupby` and `.describe` to make grouped calculations of statistical descriptions of `grad_100_value` by `level` and `control`. Which institutions appear to have the best graduation rates?

```python
# RESOURCES
# https://seaborn.pydata.org/generated/seaborn.kdeplot.html

college_completion_grouped = (college_completion
    .groupby("control")
)
sns.kdeplot(data = college_completion_grouped, x = 'grad_100_value')
```

```python
college_completion_grouped = (college_completion
    .groupby("level")
)
sns.kdeplot(data = college_completion_grouped, x = 'grad_100_value')
```

6. Create a new variable, `df['levelXcontrol']=df['level']+', '+df['control']` that interacts level and control. Make a grouped kernel density plot. Which institutions appear to have the best graduation rates?

```python
college_completion['levelXcontrol']=college_completion['level']+', '+college_completion['control']
sns.kdeplot(data=college_completion,x='grad_100_value',hue='levelXcontrol')
```

```python
college_completion.loc[:,['level','control','grad_100_value']].groupby(['level','control']).describe()
```

```python
print("It seems that just based on the mean graduation rates, private not-for-profit institutions have the highest graduation rates. That being said the sample number for private not-for-profit is more than private-for-profit and public universities combines, so the difference in scale could possibly be skewing numbers.")
```

7. Make a kernel density plot of `aid_value`. Now group your graph by `level` and `control`. What explains the shape of the graph? Use `groupby` and `.describe` to make grouped calculations of statistical descriptions of `aid_value` by `level` and `control`.

8. Make a scatterplot of `grad_100_value` by `aid_value`. Describe what you see. Now make the same plot, grouping by `level` and then `control`. Describe what you see. For which kinds of institutions does aid seem to increase graduation rates?
